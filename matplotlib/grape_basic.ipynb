# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
# plt.show() [<matplotlib.lines.Line2D at 0x1d92fab0cd0>] 보기 싫을때
```

## Title 설정
그래프에 이름 붙이기

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프') #한국어 인코딩 실패
```

### 한글 폰트 설정

```python
import matplotlib as ml
ml.rcParams['font.family'] = 'Malgun Gothic'
ml.rcParams['font.size'] = 18
ml.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

사용 가능한 폰트 목록 확인하기

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist]
```
