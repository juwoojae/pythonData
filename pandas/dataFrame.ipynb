2차원 데이터 (Series 들의 모음)

## Data 준비
사전 (dict) 자료구조를 통해 생성

예) 슬램덩크 주요 인물 8명에 대한 데이터

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
data
```

```python
data['이름']
```

```python
data['키']
```

## dataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data) #DataFrame 객체를 위에서 만들어준 data dict 을 넣어준다.
```

```python
df
```

## 데이터 접근

```python
df['이름']
```

```python
df['키']
```

```python
df[['이름', '키']]
```

## DataFrame 객체 생성 (Index 지정)

```python
df = pd.DataFrame(data, index = ['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df
```

## DataFrame 객체 생성 (Column 지정)
data 중에서 원하는 column 만 선택하거나, 순서 변경 가능

```python
df = pd.DataFrame(data, columns=['이름', '학교', '키'])
df
```

```python
df = pd.DataFrame(data, columns=['이름', '키', '학교'])
df
```
