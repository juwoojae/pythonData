# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd
data = {
    '이름': ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교': ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키': [197, 184, 168, 187, 188, 202, 188, 190],
    '국어': [90, 40, 80, 40, 15, 80, 55, 100],
    '영어': [85, 35, 75, 60, 20, 100, 65, 85],
    '수학': [100, 50, 70, 70, 10, 95, 45, 90],
    '과학': [95, 55, 80, 75, 35, 85, 40, 95],
    '사회': [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기': ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df.index.name = '지원번호'
df
```

```python
df.to_csv('score.csv', encoding='utf-8-sig') # excel 로 열었을때 인코딩 하는 방법
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False) # 인덱스를 제외하는 방법
```

### 텍스트(.txt) 파일로 저장

```python
df.to_csv('score.txt', sep='\t') # tab 으로 구분된 텍스트 파일 생성
```

### 엑셀 파일로 저장

```python
df.to_excel('score.xlsx')
```

## 열기
csv 파일 열기

```python
df= pd.read_csv('score.csv') # read_csv() 의 리턴값 = DataFrame 이다.
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 무시하고 싶은 row 제외 가능
# 지정된 갯수 만큼의 row 를 건너뛴다.
# 이름 학교 키 국어 영어 수학 과학 사회 sw특기도 하나의 rows 로 간주된다.
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1,3,5]) # 1,3,5 rows 제외하기
df
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 갯수 만큼의 row 만 가져온다.
df
```

```python
df = pd.read_csv('score.csv',skiprows=2 , nrows=4) # 앞의 2개 제외이후, 4 row 를 가지고 온다.
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호') # 인덱스 설정하기
df
```
