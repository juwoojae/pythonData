# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정 (약간 쿼리에서 update 문인가?)

```python
df['학교'].replace({'북산고': '상북고', '능남고': '무슨고'}) # 딕셔너리 형태로 바꾸고자 하는 행과 바꾸려고 하는 값 넣어주기
```

```python
df['학교'] = df['학교'].replace({'북산고': '상북고'}) # 딕셔너리 형태로 바꾸고자 하는 행과 바꾸려고 하는 값 넣어주기 inplace 사용하지 말기.
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
```

```python
df
```

```python
df['결과'] = 'Fail' # 결과 Column 을 추가하고 전체 데이터는 Fail 로 초기화
df
```

```python
df.loc['1번']
```

```python
df.loc[df['총합'] > 400 ,'결과'] = 'Pass' # loc[row , col]
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 Column 삭제하기
```

```python
df.drop(columns=['국어', '영어', '수학'], inplace= True) # 국어, 영어, 수학 column 을 삭제 (이건 인플레이스 해도 됨)
df
```

## Row 삭제

```python
df.drop(index='4번') # 4번 학생 데이터 row 삭제
```

```python
filt = df['수학'] < 80 # 수학 점수 80 점 미만 학생 필터링
filt
df[filt]
```

```python
df[filt].index # logical index 에서 걸러진 나머지 인덱스들이 나온다.
```

```python
df.drop(index=df[filt].index)
```
