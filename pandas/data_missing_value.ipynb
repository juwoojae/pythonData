## 데이터 제외하기 dropna


```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
df.dropna(inplace=True) # nan 가 존재하는 행은 무조건 삭제
df
```

* axis : index or column 열을 지울것인가, 컬럼을 지울것인가.
* how : any or all 하나라도 있으면, 모두가 NaN 일때.

```python
df.dropna(axis='index', how='any')
```
