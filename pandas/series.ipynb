# 1. Series
1 차원 데이터 (정수, 실수, 문자열등)

## Series 객체 생성
예) 1월부터 4월까지 평균 온도 데이터 (-20, -10, 10, 20)

```python
temp = pd.Series([-20, -10, 10, 20])
temp # 주피터 노트북의 경우 print 를 해주지 않아도 출력이 된다.
```

```python
temp[0] # 1월 온도
```

## Series 객체 생성 (index 지정)

```python
temp = pd.Series([-20, -10, 10, 20], index = ['Jan', 'Feb', 'Mar', 'Apr'])
temp
```

```python
temp['Jan'] #1월 온도 
```
