# 11. 데이터 정렬

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('키') # 키 기준으로 오름차순 정렬
```

```python
df.sort_values('키', ascending=False) # 키 기준으로 내림차순 정렬
```

```python
df.sort_values(['수학', '영어'], ascending=False)  # 1. 수학기준 내림차순 2. 영어 기준 내림차순 (만약 수학과 영어가 같은경우 영어 내림차순)
```

```python
df.sort_values(['수학', '영어'])
```

```python
df.sort_values(['수학', '영어'], ascending=[True, False])
```

```python
df['키'].sort_values() # 키 데이터를 가지고 옴(series) 에 대해서 오름차순 정렬
```

```python
df.sort_index()
```
